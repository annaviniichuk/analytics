# Did Sales Training Work? 

**Date:** March 2026 &nbsp;&nbsp;|&nbsp;&nbsp; **Team:** Data Science

---

> **TL;DR:** Employees who attended the training program sold **~19 more units in Year 2** — a **44% lift** over what they would have achieved without it. The effect is statistically significant.

### The business question

Each year, the company offers a sales training program to employees who hit **40 or more sales** in their first year — the idea being that high performers can be pushed even further with the right support.

But does the training actually drive results? Or do those employees just keep doing what they were already doing?

This report uses **1,000 employee records** across two years to answer that question with data.

### How the program works

The eligibility rule is straightforward: employees who sold **40 or more units in Year 1** automatically qualify for training. Everyone below that threshold does not.

### Why a simple comparison doesn't work

At first glance, you might compare the Year 2 sales of the trained group vs. the untrained group and call it a day. The problem: **the trained group was already stronger performers to begin with** — that's how they qualified. Any comparison between the two groups would mix together the effect of training with the effect of already being a high performer, making the training look better than it really is.

We need a smarter approach.

### Metrics and findings

The key insight: an employee who sold **39 units** vs. one who sold **41 units** are essentially the same kind of performer — nearly identical in skill, motivation, and trajectory. The only meaningful difference between them is that one just barely qualified for training and one just barely didn't.

By focusing on employees **within 10 units of the cutoff** (30–50 sales), we create a natural experiment: two nearly-identical groups, separated only by whether they received training. The gap in their Year 2 performance is our best estimate of the training's true effect.

The chart below shows this clearly — look at what happens right at the cutoff line.

The jump at the cutoff is the training effect — the extra performance attributable to the program, not to the employee's prior ability.

| Metric | Value |
|--------|-------|
| **Absolute effect** | +19 additional sales in Year 2 |
| **Relative lift** | +44% above the untrained baseline |
| **Statistical confidence** | p < 0.001 (extremely unlikely to be chance) |
| **Sample used** | 519 employees within ±10 units of the cutoff |

## Model Validation

**Check 1: Are the two groups actually comparable?**

The whole approach depends on employees just below and just above the 40-unit cutoff being essentially interchangeable — 
same mix of full-time and part-time workers, same underlying characteristics. If the groups were systematically different, 
the gap in Year 2 sales might reflect those differences rather than the training itself.

Across the full employee base, part-time workers are overrepresented in the untrained group (28% vs. 0%). 
That's expected — part-timers tend to sell less and therefore cluster below the cutoff. But when we narrow the lens to employees 
**within 10 units of the cutoff**, part-time representation drops to **0% in both groups**. 
The two groups are identical on this dimension, supporting the assumption that they're comparable. ✓

**Check 2: Does the jump only appear at the real cutoff?**

If the training is genuinely driving the result, we should only see a performance jump at the real eligibility threshold (40 units). 
We tested this by running the same analysis at a **fake cutoff of 30 units** — a point where no program decision was made 
and no training was triggered.

Result: no meaningful jump. The estimated effect at the placebo cutoff was **+0.4 units** (p = 0.70) — 
statistically indistinguishable from zero. This tells us the pattern we found at 40 units only shows up where the training actually happened. ✓

| Validation Check | Result |
|------------------|--------|
| Part-time balance within bandwidth | 0% part-time in both groups — groups are comparable |
| Placebo test at cutoff = 30 | +0.4 units, p = 0.70 — no spurious effect detected |

## Business Impact & Recommendation

For every **100 employees** near the eligibility threshold who receive training, the program is estimated to generate approximately **1,900 additional sales units in Year 2** compared to what those same employees would have produced without it.

The effect is large, consistent, and statistically robust. This is not noise.

**Recommendation:** Continue — and consider expanding — the training program. The 40-unit cutoff creates a clean eligibility rule, and the data shows it's selecting employees who respond meaningfully to the investment. If budget allows, this analysis also raises the question of whether employees *below* the cutoff might benefit from a modified version of the program.

*Analysis based on 1,000 employee records. Method: Regression Discontinuity Design with ±10 unit bandwidth around the eligibility cutoff.*
